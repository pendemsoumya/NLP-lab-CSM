```python
import spacy
from spacy.lang.hi import Hindi
texth="ACE इंजीनियरिंग कॉलेज में शैक्षिक जीवन जीवंत है, जहां सीखने के लिए सच्ची शिक्षा दी जाती है। ACE में ज्ञान हस्तांतरण प्रक्रिया के मुख्य तत्व हैं एडुटेनमेंट और इन्फोटेनमेंट। पहला इसका मतलब है उद्यमिता के तरीके से शिक्षा, जबकि दूसरा इसका मतलब है मनोरंजन के तरीके से जानकारी साझा करना। कक्षा में बातचीत को जीवंत बनाया जाता है, जहां मूलभूत सिद्धांतों को वास्तविक जीवन के केस अध्ययन के साथ कवर किया जाता है, ताकि छात्रों को सैद्धांतिक ज्ञान (क्या पहलू) की अनुप्रयोगीय समझ विकसित हो सके। इसके बाद छात्रों को प्रयोगशालाओं में अभ्यास कराया जाता है (कैसे पहलू)। छोटे समूहों में छात्र एक-दूसरे से चर्चा करके यह समझते हैं कि क्यों और कब परियोजनाओं पर काम किया जाए, जो सेमेस्टर के अंत में किए जाते हैं।"
nlp=Hindi()
nlp.add_pipe('sentencizer')
doc=nlp(texth)
for word in doc:
  print(word.text)
```

```python
from nltk.stem import PorterStemmer

def stemming(data):
    ps = PorterStemmer()
    d = {}  # Define the dictionary
    for w in data.split():
        d[w] = ps.stem(w)  # Use the stemmer on each word
    return d
```

```python
stemming=stemming("Eat Eats Eaten Eating Talk Talking Join Joined Joining")
stemming
```

```python
from nltk.stem import SnowballStemmer
snow=SnowballStemmer("english")
text="Eat Eats Eaten Eating Talk Talking Join Joined Joining Drove Arise Feet"
l=text.split()
for i in l:
  print(i,'-->',snow.stem(i))
```

```python
from nltk.stem import LancasterStemmer
lanc=LancasterStemmer()
text="Eat Eats Eaten Eating Talk Talking Join Joined Joining Drove Arise Feet"
j=text.split()
for i in j:
  print(i,'-->',lanc.stem(i))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
st=WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('omw-1.4')
text="Eat Eats Eaten Eating Talk Talking Join Joined Joining Drove Arise Feet"
l=text.split()
for i in l:
  print(i,' | ',st.lemmatize(i))
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
doc=nlp("Eat Eats Eaten Eating Talk Talking Join Joined Joining Drove Arise Feet")
for i in doc:
  print(i," | ",i.lemma_)
```

```python
nltk.download('tagsets')
pos_f_f=nltk.data.load('help/tagsets/upenn_tagset.pickle')
for word , tag in pos:
  f_f=pos_f_f.get(tag,"Unknown POS Tag")
  print(f"{word} : {tag}  ({f_f})")
```

```python
def gew():
  return random.choice(w_l)
```

```python
def glw(n):
  wd=[word for word in w_l if len(word)==n]
  return random.choice(wd)
```
